```python
plt.style.use('ggplot')
```

```python
import statsmodels.api as sm
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,7))
sns.boxplot(data=df_wage, x='education', y='wage', ax=ax[0])
plt.sca(ax[0])
plt.xticks(rotation=45)
plt.xlabel('Education')
plt.ylabel('Wage')

sns.boxplot(data=df_wage, x='year', y = 'wage', ax=ax[1], color='lightblue')
plt.sca(ax[1])
plt.xlabel('Year')
plt.ylabel('Wage')

plt.sca(ax[2])
plt.scatter(df_wage['age'], df_wage['wage'], color='lightgreen')
plt.xlabel('Age')
plt.ylabel('Wage')
lowess = sm.nonparametric.lowess(df_wage['wage'], df_wage['age'], frac=0.2)
ax[2].plot(lowess[:,0], lowess[:,1])

```
